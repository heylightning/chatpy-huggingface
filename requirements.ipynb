## You will need the following libraries to get started:
* **transformers**
* **streamlit**
* **PyPDF2**
* **pyautogui**

For **transformers**: -
`
    pip install transformers
`

For **streamlit**: -
`
    pip install streamlit
`

For **PyPDF2**: - 
`
    pip install PyPDF2
`

For **pyautogui**: -
`
    pip install pyautogui
`

Look for **prediction_prototype.ipnb** for viewing how the model works and outputs the result.

*Note: These requirements are staged for version 1.0.0.*
